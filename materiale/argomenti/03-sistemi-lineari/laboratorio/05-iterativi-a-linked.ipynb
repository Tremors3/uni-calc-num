## Metodi di decomposizione

Ogni matrice può essere scritta come la somma di una diagonale e due triangolari.

$$A = \begin{pmatrix}
\color{red}{a_{11}}&\color{green}{a_{12}}&\color{green}{a_{13}} &\color{green}{a_{14}} &\color{green}{\cdots}  &\color{green}{\cdots}   &\color{green}{a_{1n}}  \\
{\color{blue}{a_{21}}}&\color{red}{a_{22}}&\color{green}{a_{23}} &\color{green}{a_{24}} &\color{green}{\cdots} &\color{green}{\cdots} &\color{green}{a_{2n}}   \\
\color{blue}{a_{31}}&\color{blue}{a_{32}} &\color{red}{a_{33}}&\color{green}{a_{34}} & & & \color{green}{a_{3n}}   \\
 \color{blue}{\cdots} & \color{blue}{\cdots}&   &\color{red}{\ddots}&\color{green}{\cdots}& \color{green}{\cdots}& \\
\color{blue}{a_{i1}}& \color{blue}{\cdots}& &\color{blue}{a_{ii-1}}&\color{red}{a_{ii}}&\color{green}{a_{ii+1}} &\color{green}{a_{in}} \\
 \color{blue}{\cdots} &\color{blue}{\cdots} &\color{blue}{\cdots}      && &\color{red}{\ddots}  \\
\color{blue}{a_{n-11}}  &\color{blue}{\cdots} &\color{blue}{\cdots}&\color{blue}{\cdots} & \color{blue}{{a_{n-1n-2}}}&\color{red}{a_{n-1n-1}}&\color{green}{a_{n-1n}}\\
\color{blue}{a_{n1}}  & \color{blue}{\cdots}&  \color{blue}{\cdots}&\color{blue}{\cdots} &\color{blue}{\cdots} &\color{blue}{ {a_{nn-1}}}    &\color{red}{a_{nn}}
\end{pmatrix}
$$

$$E = \begin{pmatrix}
0&0&0 &0 &{\cdots}  &0  &0 \\
-{\color{blue}{a_{21}}}&0&0 &0 &{\cdots} &{\cdots} &0  \\
-\color{blue}{a_{31}}&-\color{blue}{a_{32}} &0&0& & & 0   \\

-\color{blue}{a_{i1}}& \color{blue}{\cdots}& &-\color{blue}{a_{ii-1}}&0&0 &0\\
 \color{blue}{\cdots} &\color{blue}{\cdots} &\color{blue}{\cdots}      && &0  \\
-\color{blue}{a_{n-11}}  &\color{blue}{\cdots} &\color{blue}{\cdots}&\color{blue}{\cdots} & -\color{blue}{{a_{n-1n-2}}}&0&0\\

$$F=\begin{pmatrix}
0&-\color{green}{a_{12}}&\color{green}{a_{13}} &-\color{green}{a_{14}} &\color{green}{\cdots}  &\color{green}{\cdots}   &-\color{green}{a_{1n}}  \\
0&0&-\color{green}{a_{23}} &-\color{green}{a_{24}} &\color{green}{\cdots} &\color{green}{\cdots} &-\color{green}{a_{2n}}   \\

0  &0 &0&{\cdots} & 0&0&-\color{green}{a_{n-1n}}\\
0 & 0&  {\cdots}&{\cdots} &{\cdots} &0    &0
\end{pmatrix}$$


$$D = \begin{pmatrix}
\color{red}{a_{11}}&0&0 &0 & {\cdots}  & {\cdots}   &0\\
0&\color{red}{a_{22}}&0 & 0& {\cdots} & {\cdots} &0 \\

\end{pmatrix} $$



$$A = D-E-F $$

## Metodo di Jacobi

Decomposizione della matrice $A=D-(E+F)$

$$Dx^{(k+1)} = (E+F)x^{(k)} + b $$
per ottenere *formalmente* la matrice di iterazione si può riformulare come

$$x^{(k+1)} = \underbrace{D^{-1}(E+F)}_{\mathcal J}x^{(k)} + \underbrace{D^{-1}b}_{c}  $$

Formula effettivamente implementata:

$$ x^{(k+1)} = D^{-1}((E+F)x^{(k)} + b) $$

Formalmente vediamo la matrice inversa, ma in realtà non vengono mai calcolate.
Viene infatti effettuata la divisione per la diagonale $d$ (*vedi codice*).

L'implementazione del metodo non richiede il calcolo della matrice di iterazione, ma il nuovo iterato si calcola risolvendo il sistema diagonale con matrice dei coefficienti $D$ e termine noto $(E+F)x^{(k)}+b$.

```python
import numpy as np

def jacobi(A, b, x0, tol=1e-10, Nmax=100):
    """
    Metodo iterativo di Jacobi per risolvere Ax = b.

    Parametri:
        A    : matrice dei coefficienti (n x n)
        b    : termine noto (n,)
        x0   : stima iniziale (n,)
        tol  : tolleranza per il criterio d'arresto
        Nmax : numero massimo di iterazioni

    Ritorna:
        x    : soluzione approssimata
    """

    x = x0.copy()                   # iterata corrente x^(k)
    d = np.diag(A)                  # diagonale di A

    # Controllo: elementi diagonali non nulli
    if np.any(d == 0):
        raise ValueError("La matrice ha elementi diagonali nulli.")

    # Parte fuori diagonale: E + F
    E_F = A - np.diag(d)

    for k in range(Nmax):

        # Formula di Jacobi:
        # x^(k+1) = D^{-1} (b - (E+F)x^(k))
        x_new = (b - E_F @ x) / d

        # Criterio d'arresto (norma assoluta)
        if np.linalg.norm(x_new - x) < tol:
            print(f"Convergenza in {k} iterazioni.")
            return x_new

        x = x_new.copy()

    print("Warning: numero massimo di iterazioni raggiunto!")
    return x
```

```python
# Gli elementi diagonali sono sicuramente maggiori della somma degli altri elementi.
# La matrice A rispetta le regole di convergenza dei due metodi di Jacobi e Gauss-Seidel

# --- Sistema di test ---
A = np.array([
    [4.0, 1.0, 2.0],
    [3.0, 5.0, 1.0],
    [1.0, 1.0, 3.0]
])

b = np.array([4.0, 7.0, 3.0])

# soluzione esatta: [0.5, 1.0, 0.5]
x0 = np.zeros_like(b)

# --- Risoluzione ---
sol = jacobi(A, b, x0)

print("Soluzione:", sol)
print("Residuo:", np.linalg.norm(A @ sol - b))
```

## Metodo di Gauss-Seidel

Decomposizione della matrice

$$A = (D-E) - F $$

$$ (D-E)x^{(k+1)} = Fx^{(k)} + b$$
per ottenere *formalmente* la matrice di iterazione si può riformulare come
$$x^{(k+1)} = \underbrace{(D-E)^{-1}F}_{\mathcal G}x^{(k)} + \underbrace{(D-E)^{-1}b}_{c} $$

L'implementazione del metodo non richiede il calcolo della matrice di iterazione, ma la nuova iterata viene calcolata utilizzando il metodo di sostituzione in avanti applicato al sistema con matrice $(D-E)$ e termine noto $Fx^{(k)}+b$.

```python
import numpy as np
from scipy.linalg import solve_triangular

def gauss_seidel(A, b, x0, tol=1e-10, Nmax=100):
    """
    Metodo di Gauss-Seidel per la soluzione del sistema lineare Ax = b.

    L'idea è riscrivere il sistema nella forma iterativa:
        x^{(k+1)} = (D - E)^{-1} (F x^{(k)} + b)

    dove:
        A = D - E - F
    con:
        D = matrice diagonale di A
        -E = parte strettamente inferiore
        -F = parte strettamente superiore
    """

    # Inizializzazione:
    # x rappresenta l'iterata corrente x^(k)
    # x_new rappresenta la nuova iterata x^(k+1)
    x = x0.copy()
    x_new = np.zeros_like(x)

    # --- Scomposizione della matrice A ---
    # D - E = parte triangolare inferiore (inclusa la diagonale)
    # F     = parte triangolare superiore (senza diagonale)
    D_E = np.tril(A)        # matrice triangolare inferiore (D - E)
    F   = -np.triu(A, 1)    # matrice triangolare superiore (F)

    # Ciclo iterativo
    for k in range(Nmax):

        # Ad ogni iterazione dobbiamo risolvere un sistema lineare triangolare:
        #
        # (D - E) x^(k+1) = F x^(k) + b
        #
        # Questo è possibile perché (D - E) è triangolare inferiore.
        #
        # solve_triangular risolve direttamente questo sistema senza inversione.
        x_new = solve_triangular(
            D_E,
            F @ x + b,
            lower=True
        )

        # Criterio di arresto basato sulla differenza tra iterate successive
        if np.linalg.norm(x_new - x) < tol:
            print(f"Convergenza in {k} iterazioni.")
            return x_new

        # Aggiornamento dell'iterata
        x = x_new.copy()

    # Se si raggiunge Nmax senza convergenza
    print("Warning: numero massimo di iterazioni raggiunto!")
    return x_new
```

```python
# --- Sistema di test ---
A = np.array([
    [4.0, 1.0, 2.0],
    [3.0, 5.0, 1.0],
    [1.0, 1.0, 3.0]
])

b = np.array([4.0, 7.0, 3.0])

x0 = np.zeros_like(b)

# --- Risoluzione ---
sol = gauss_seidel(A, b, x0)

print("Soluzione:", sol)
print("Residuo:", np.linalg.norm(A @ sol - b))
```
