#Metodo di Jacobi
Decomposizione della matrice $A=D-(E+F)$

$$Dx^{(k+1)} = (E+F)x^{(k)} + b $$
per ottenere *formalmente* la matrice di iterazione si può riformulare come

$$x^{(k+1)} = \underbrace{D^{-1}(E+F)}_{\mathcal J}x^{(k)} + \underbrace{D^{-1}b}_{c}  $$

L'implementazione del metodo non richiede il calcolo della matrice di iterazione, ma il nuovo iterato si calcola risolvendo il sistema diagonale con matrice dei coefficienti $D$ e termine noto $(E+F)x^{(k)}+b$

#Metodo di Gauss-Seidel
Decomposizione della matrice
$$A = (D-E) - F $$

$$ (D-E)x^{(k+1)} = Fx^{(k)} + b$$
per ottenere *formalmente * la matrice di iterazione si può riformulare come
$$x^{(k+1)} = \underbrace{(D-E)^{-1}F}_{\mathcal G}x^{(k)} + \underbrace{(D-E)^{-1}b}_{c} $$

L'implementazione del metodo non richiede il calcolo della matrice di iterazione, ma la nuova iterata viene calcolata utilizzando il metodo di sostituzione in avanti applicato al sistema con matrice $(D-E)$ e termine noto $Fx^{(k)}+b$.
