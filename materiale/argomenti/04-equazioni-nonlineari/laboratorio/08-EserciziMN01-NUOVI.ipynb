## Esercizio 13

Creare il grafico delle curve parametriche definite dalle seguenti formule
1. $t\in [-1,1]$
$$\begin{cases} x(t)&=&\sqrt{1 - t^2} + \sqrt{|t|}\\
y(t) &=&-\sqrt{1 - t^2} + \sqrt{|x_h|}
\end{cases}$$
2. $t\in[0,2\pi]$
$$\begin{cases}
x(t)&=&2\cos(t)\\
y(t)&=& 3\sin(t)(1 + 0.2\sin(t))\end{cases}$$
3. $t\in[0,2\pi]$
$$\begin{cases}
x(t) &=& r(t) \cos(t)\\
y(t) &=& r(t) \sin(t)
\end{cases}$$
dove $r(t) = \cos(2t)$
4. $t\in[-2,2]$
$$\begin{cases}
x(t) &=& t\\
y(t) &=& |t| - |t-0.5| - |t+0.5| + 1
\end{cases}$$
5. $t\in [0,12\pi]$
$$\begin{cases}
x(t) &=& r(t)\sin(t)\\
y(t) &=& r(t)\cos(t)
\end{cases}$$
dove $r(t) = e^{\cos(t)} - 2\cos(4t) + (\sin(t/12))^5$

*Suggerimento*: NO cicli
